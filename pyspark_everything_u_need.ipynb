```python
df_pyspark = spark.read.csv('test1.csv',header=True,inferSchema=True)
```

```python
df_pyspark.na.drop().show() #the whole rows will be deleted 
```

```python
# any == how
df_pyspark.na.drop(how="any").show() # if it see a row with null everthing will be droped on that row 
```

```python
df_pyspark.na.drop(how="all").show() # only drop a drop a row if you have all  the values of row as null 
```

```python
##threshold

df_pyspark.na.drop(how="any",thresh=2).show() # check for rows where non null values are atleast 2 if it is less than 2 non null values then remove that row 
```

```python
#subset  the coloms that have null values in it corresponing rows values are deleted mostly tthe whoel row 

df_pyspark.na.drop(how="any",subset=['Motor1 PWM']).show() 
```

```python
df_pyspark.na.drop(how="any",subset=['Motor3 PWM']).show() 
```

```python
 df_pyspark.na.fill("Missing",['sample index','Motor1 PWM']).show() #if the type of the colom is string it will replace NULL with Missing word and it will only do to particular cloum u sepcify if u dont then u have to keep it empty 
```

```python
df_pyspark.na.fill(0).show()
```

```python
df_pyspark.show()
```

```python
# impute mean values to rh NULL vales to make it work 
from pyspark.ml.feature import Imputer

imputer = Imputer(inputCols=['Battery voltage (V)','Motor1 PWM','Motor2 PWM','Motor3 PWM','Motor4 PWM'],
                 outputCols=[ "{}_imputed".format(c)for c in ['Battery voltage (V)','Motor1 PWM','Motor2 PWM','Motor3 PWM','Motor4 PWM']]).setStrategy("mean")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
